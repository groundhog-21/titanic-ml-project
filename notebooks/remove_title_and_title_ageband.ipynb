```python
import pandas as pd

# Load data
train_v3 = pd.read_csv("../data/processed_v3/train_features.csv")
test_v3  = pd.read_csv("../data/processed_v3/test_features.csv")

# Columns to remove
cols_to_drop = ["Title", "Title_AgeBand"]

for col in cols_to_drop:
    if col in train_v3.columns:
        print(f"Removing {col} from train")
        train_v3 = train_v3.drop(columns=[col])
    if col in test_v3.columns:
        print(f"Removing {col} from test")
        test_v3 = test_v3.drop(columns=[col])

# Save updated data
train_v3.to_csv("../data/processed_v3/train_features.csv", index=False)
test_v3.to_csv("../data/processed_v3/test_features.csv", index=False)
```

```python
import json

meta_path = "../data/processed_v3/processed_metadata.json"

with open(meta_path) as f:
    meta_v3 = json.load(f)

for key in ["Title", "Title_AgeBand"]:
    if key in meta_v3:
        print(f"Removing {key} from metadata")
        del meta_v3[key]

with open(meta_path, "w") as f:
    json.dump(meta_v3, f, indent=4)

meta_v3
```
